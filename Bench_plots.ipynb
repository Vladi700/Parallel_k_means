```python
# 1) Standard imports
import pathlib
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 2) Point this to the directory where you put the .pkl files
PKL_DIR = pathlib.Path("/Users/test/Desktop/Parallel_k_means/benchmark_result").expanduser()

# 3) Load every pickle into a list of dicts
rows = []
for pkl_file in PKL_DIR.glob("*.pkl"):
    with pkl_file.open("rb") as f:
        rows.append(pickle.load(f))

# 4) One big DataFrame
df = pd.DataFrame(rows)
print(df.head())
print(df.columns)

# ────────────────────────────────────────────────────────────────
#  Example plots  (feel free to add / remove as you need)
#  ────────────────────────────────────────────────────────────────

# A) Runtime vs. chunk size, coloured by #workers
plt.figure()
pivot = df.pivot_table(index="chunks", columns="workers",
                       values="kmeans_mean_time")   # change to your column names
pivot.plot(marker="o")
plt.ylabel("Mean runtime (s)")
plt.title("k-means|| runtime vs. chunk size")
plt.grid(True)
plt.show()

# B) Speed-up curve for a chosen chunk size (e.g. 16)
chosen_chunk = 16
sub = df[df["chunks"] == chosen_chunk]

plt.figure()
plt.plot(sub["workers"], sub["kmeans_mean_time"], marker="o")
plt.gca().invert_xaxis()       # lower runtime is better
plt.xlabel("Workers")
plt.ylabel("Mean runtime (s)")
plt.title(f"Speed-up (chunks={chosen_chunk})")
plt.grid(True)
plt.show()

# C) Memory usage (if you recorded it) vs. threads
if "memory_used" in df.columns:
    plt.figure()
    plt.scatter(df["threads"], df["memory_used"]/1e9)
    plt.xlabel("Threads per worker")
    plt.ylabel("Memory used (GB)")
    plt.title("Memory vs. threads")
    plt.grid(True)
    plt.show()
```

```python
heat = (
    df.pivot_table(index="workers",
                   columns="chunks",
                   values="kmeans_mean_time")      # <-- change if your column name differs
      .sort_index()                           # keep workers ascending (1,2,3…)
      .sort_index(axis=1)                     # keep chunks  ascending (4,6,8…)
)

# ----------------------------------------------------------
# 2) Plot
# ----------------------------------------------------------
fig, ax = plt.subplots(figsize=(8, 4))
im = ax.imshow(heat.values, cmap="viridis", origin="lower")

# axis tick labels
ax.set_xticks(range(len(heat.columns)), heat.columns)
ax.set_yticks(range(len(heat.index)),  heat.index)
ax.set_xlabel("Chunk size")
ax.set_ylabel("Number of workers")
ax.set_title("Mean runtime (s)")

# put the numbers on each cell
for (i, j), val in np.ndenumerate(heat.values):
    ax.text(j, i, f"{val:.1f}", ha="center", va="center", color="white" if val<im.norm(heat.values).max()/2 else "black", fontsize=8)

# color bar
cbar = fig.colorbar(im, ax=ax)
cbar.set_label("Seconds")

plt.tight_layout()
plt.show()
```
